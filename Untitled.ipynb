```python
# Female Chinstrap종의 bill_length 높은순 정렬
cond1 = df_copy['sex'] == 'Female'
cond2 = df_copy['species'] == 'Chinstrap'

df_copy.loc[cond1 & cond2, : ].sort_values('bill_length_mm', ascending =False).head(10)
```
