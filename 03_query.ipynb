## practice

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
# - 남성 흡연자 조건 필터링
# - tip을 가장 많이 낸 순서로 내림차순 정렬
# - 상위 10개 출력
```

```python
cond1 = tips['sex'] == 'Male'

cond2 = tips['smoker'] == 'Yes'

# tips_man = tips.loc[cond1 & cond2]
# tips_man.sort_values('tip', ascending = False).head(10) 위에 처럼 복잡하게 하지 않고 밑에 바로 sort_values를 붙일 수 있음
tips.loc[cond1 & cond2].sort_values('tip', ascending=False).head(10)
```

```python
# - 식사인원이 5명 이상인 저녁테이블
# - 흡연자가 없는 주말 테이블
# - total_bill, tip 출력
```

```python
cond1 = tips['size'] >= 5
cond2 = tips['time'] == 'Dinner'
cond3 = tips['smoker'] == 'No'
cond4 = tips['day'].isin(['Sat', 'Sun'])

tips.loc[(cond1 & cond2) & (cond3 & cond4), 'total_bill':'tip']
tips.loc[(cond1 & cond2) & (cond3 & cond4), ['total_bill', 'tip']]


```
