## 자전거 데이터 활용

```python
df = pd.read_csv('data/seoul_bicycle.csv')
df.head(2)
```

```python
df.info()
```

```python
df_copy = df.copy()
```

```python
df_copy['대여일자'] = pd.to_datetime(df_copy['대여일자'], format = '%b-%d-%Y') # 우리가 갖고있는 글자 객체를 파이썬이 읽을 수 있는 날짜로 바꿔줌
df_copy.info()
```

```python
cond = df_copy['대여일자'].dt.dayofweek == 3
df_copy.loc[cond]
```

```python
df_copy.head(3)
```

```python
df_copy['운동량'] = pd.to_numeric(df_copy['운동량'], errors = 'coerce') # 에러가 난 부분을 Nan으로 바꾸기
```

```python
df_copy.loc[2344]
```

### cut()

```python
bins = [0, 100, 200, 300, df_copy['운동량'].max()]
pd.cut(df_copy['운동량'], bins).value_counts()
```

```python
pd.cut(df_copy['운동량'], bins =10)
pd.cut(df_copy['운동량'], bins =10).value_counts()
```

```python
pd.qcut(df_copy['운동량'], q=10)
pd.qcut(df_copy['운동량'], q=10).value_counts() # qcut은 비율대로 나누기
```

```python
qcut_bins = [0, 0.2, 0.8, 1] # 하위 20프로 상위 20프로 
df_copy['운동량_한글'] = pd.qcut(df_copy['운동량'], qcut_bins, labels = ['적음', '보통', '믾음'])
df_copy.head(2)
```
